````python
text_1 = '人生好难啊' '
s_1 = SnowNLP(text_1)
text_2 = '人生好难啊，要怎样才能每天开心的活着呢？'
s_2 = SnowNLP(text_1)
print(s_1.sentiments)
print(s_2.sentiments)
````

````python
text_3 = SnowNLP(u'时间不够用，每天还是努力学习。也许我是傻蛋，但我有一天会成功')
sent = text_3.sentences
for sen in sent:
    s_3 = SnowNLP(sen)
    print(s_3.sentiments)
````
